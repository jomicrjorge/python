<h2 id="objecitve">Objective of the War Card Game</h2>

The War card game is a simple card game where the card with the highest value wins. In this version of the game, we only use 2 players, with the following rules:

<ul>
  <li>Deal one card to each player.</li>
  <li>Both players reveal their card, and the player with the higher card wins the round.</li>
  <li>If there is a tie, each player receives an additional card to determine the winner.</li>
  <li>Cards that have been dealt to the players are not returned to the deck until the game is finished. This guarantees that the game cannot end in a draw.</li>
</ul>

<h2 id="deck">Create a Deck</h2>

As everyone knows, a standard deck of 52 cards has 4 suits with 13 ranks each (from the numbers 2 to 10 plus 4 face cards). To represent this in code, we could create a list/array and manually add all 52 cards one by one.

However, doing that would take a lot of time. To make our work easier, we use loops in Python. In this case, we will use a <b>for</b> loop. We can use two different methods with a for loop to create our deck:

But in order to use the for loop, we first need to create two arrays (lists): one for the suits and one for the ranks of the cards. Using these, the loop will generate the full deck for us.

To store this new list of cards, we need a place to keep the information. For this purpose, we first create an empty array/list and then, using the <code>append</code> function, we store each card inside our <b>deck</b> variable.

<ol>
  <li>Using the <code>append</code> function</li>
  <li>Using an f-string inside a list comprehension</li>
</ol>

<h3>1 – Using the <code>append</code> Function</h3>


```python
figures = [2, 3, 4, 5, 6, 7, 8, 9, 10, "Ace", "King", "Queen", "Jack"]
naipes = ["clubs", "diamonds", "hearts", "spades"]

# We need to define an empty list so that we can append our cards inside the for loop
deck = []

for fig in figures:
    for naipe in naipes:
        # In this case we only use fig because we only need the rank (figure) of the card.
        # For this game, the suit (naipe) is not important.
        deck.append(fig)

print(deck)# This code isn't necessary
print(len(deck))# This code isn't necessary

```
